```python
import pandas as pd
import anthropic
```

```python
youtube = pd.read_csv('Youtube.csv')
youtube.head()
```

```python
client = anthropic.Anthropic(
    # defaults to os.environ.get("ANTHROPIC_API_KEY")
    api_key="",
)
def get_category(prompt):
    message = client.messages.create(
        model="claude-3-opus-20240229",
        max_tokens=2008,
        temperature=0,
        system="You are an AI assistant trained to sort comments on videos relate to dark matter - concept in astrophysics into predefined categories. Your goal is to analyze each comment and assign the most relevant category to each set.\nFirst set:\n1. Theories and Hypotheses\n2. Observations and Evidence\n3. Questions and Clarificatxqdions\n4. Analogies and Metaphors\n5. Personal Opinions\n6. Philosophical or Religious Perspectives\n7. Jokes and Puns\n8. Satirical Comments\n9. Informative Comments\n10. Explanatory Remarks\n11. References to Additional Resources\n12. Unrelated Personal Anecdotes\n13. Comments on Video Production or Presentation\n14. Advertisements\n15. Self-promotion\n16. Insults and Personal Attacks\n17. Controversial or Provocative Statements\nSecond set \n1. Appreciation for the Content\n2. Encouragement for Further Discussions\n3. Criticism of the Explanations\n4. Discouragement or Pessimism\n5. Statements without Clear Sentiment\n6. Mixed Opinions\nFor each comment, first output the index of the comments follow by a semicolon, then output the index of the comment's category in the first set, followed by a semicolon, then the index of the comment's category in the second set. Do not include any other text or explanations.",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt
                    }
                ]
            }
        ]
    )
    return message.content
```

```python
comments = youtube['comment_text']
#exact 30 comments from each iteration
from tqdm import tqdm
for i in tqdm(range(0, len(comments), 30)):
    prompt = ''
    for j in range(len(comments[i:i+30])):
        prompt += str(j+1 + i)  + '. ' + comments[i:i+30][i + j] + str("\\n")
    # print(prompt)

    # print(prompt)[0].text[:]
    try:
        result = get_category(prompt)
    except Exception as e:
        print(e)
        continue
    #result = result[0].text[:]
#save the result to a txt file
    try:
        result = result[0].text[:]
    except:
        pass
    with open('result_oups.txt', 'a') as f:  # Use 'a' mode instead of 'w' to append to the file
        lines = result.split('\n')
        if len(lines) == 1:
            lines = result.split(' ')
        if len(lines) > 30:
            print('Error', len(lines), lines)
        for line in lines:
            f.write(line + '\n')
        f.close()
```
